```python
import numpy as np
import matplotlib.pyplot as plt
```

## 线性模型
MSE: mean-square error 平均平方误差

这里模拟一个线性模型在[0, 4]上每隔0.1寻找最小的MSE

```python
x_data = [float(i) for i in range(1, 6)]
y_data = [2 * i for i in x_data]

def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
    print(f"w: {w}")
    l_sum = 0
    for x, y in zip(x_data, y_data):
        y_pred = forward(x)
        loss_value = loss(x, y)
        l_sum += loss_value
        print('\t', x, y, y_pred, loss_value)
    mse = l_sum / len(x_data)
    print(f"MSE: {mse}")
    w_list.append(w)
    mse_list.append(mse)
```

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
```

## 作业

- 写一个符合$\hat{y} = \omega x + b$的线性模型, 并绘制出3D的Loss Surface.

```python
def forward(x):
    return w * x + b

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

w_list = []
b_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
    for b in np.arange(-2.0, 2.1, 0.1):
        l_sum = 0
        for x, y in zip(x_data, y_data):
            l = loss(x, y)
            l_sum += l
        mse = l_sum / len(x_data)
        w_list.append(w)
        b_list.append(b)
        mse_list.append(mse)
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(w_list, b_list, mse_list, 'blue')
ax.set_title('3D curve plot')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('mse')
```
